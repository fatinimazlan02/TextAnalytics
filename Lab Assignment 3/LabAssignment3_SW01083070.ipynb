```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import pandas as pd

# Download the stopwords dataset
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')


# Define stop words
stop_words = stopwords.words('english')
more_stopwords = ['u', 'im', 'c', 'maxaxaxaxaxaxaxaxaxaxaxaxaxaxax']
stop_words.extend(more_stopwords)

# Initialize stemmer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Define the preprocessing function
def preprocess(text):
    text = text.lower()  # Lowercase text
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\b\w*\d\w*\b', '', text)  # Remove words containing numbers
    text = re.sub(r'\S+@\S+', '', text) 
    tokens = word_tokenize(text)  # Tokenize text
    tokens = [word for word in tokens if word not in stop_words]  # Remove stopwords

    tokens = [lemmatizer.lemmatize(word) for word in tokens]  # Lemmatization
    return tokens


# Apply preprocessing
data['message_clean'] = data['text'].apply(preprocess)

data.head(30)
```

```python
#perform LDA using Gensim library

#import topic modelling as well
from gensim import corpora
from gensim.models import LdaModel
```

```python
#create a Gensim Dictionary object from the preprocessed documents
dictionary = corpora.Dictionary(data['message_clean'])
#convert each preprocessed document into a bag-of-words representation using the dictionary
corpus = [dictionary.doc2bow(doc) for doc in data['message_clean']]
```

```python
lda_model = LdaModel(corpus, num_topics=4, id2word=dictionary, passes=15)
```

```python
#now see the result

#create an empty array to store dominant topic labels for each document
article_labels = []

#iterate over each processed document
for i, doc in enumerate(data['message_clean']):
    #for each document,convert to box representation
    bow = dictionary.doc2bow(doc)
    #get list of topic probabilities
    topics = lda_model.get_document_topics(bow)
    #determine topic with highest probability
    dominant_topic = max(topics, key=lambda x:x[1])[0]
    #appendf to the list
    article_labels.append(dominant_topic)
```

```python
#create dataframe to see them
df = pd.DataFrame({"Text": data['text'], "Topic": article_labels})

#Now print them to see the result
print("Table with Articles and Topic:")
print(df)
```

```python
#Print the top terms for each topic
print("Top Terms for Each Topic:")
for idx, topic in lda_model.print_topics():
    print(f"Topic {idx}:")
    terms = [term.strip() for term in topic.split("+")]
    for term in terms:
        weight, word = term.split("*")
        print(f"- {word.strip()} (weight: {weight.strip()})")
    print()
```

Topic 0 relates to sports or games, where terms like "game" and "team" have high significance, indicating their importance in this topic.

Topic 1 pertains to domestic politics, where terms like "government" and "people" are particularly significant, highlighting their importance in this topic.

Topic 2 concerns computer systems, where terms like "x," "key," and "file" are especially significant, indicating their importance in this topic.

Topic 3 involves discussions where individuals share their views on various subjects, with common words like "one," "would," and "don't" frequently used.

```python
#Print top terms for each topic
for topic_id in range(lda_model.num_topics):
    print(f"Top terms for Topic #{topic_id}:")
    top_terms = lda_model.show_topic(topic_id, topn=10)
    print([term[0] for term in top_terms])
    print()
```
